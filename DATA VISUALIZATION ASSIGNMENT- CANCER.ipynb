```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("cancer.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
## Create a bar chart or pie chart to visualize the distribution of diagnoses in the dataset. 
## This will provide insights into the balance of malignant and benign cases.
```

```python
diagnoses = {"Malignant":50, "Benign":50}
plt.figure(figsize=(8, 6))
plt.pie(diagnoses.values(), labels=diagnoses.keys(), autopct='%1.1f%%')
plt.title("Distribution of Diagnoseis i a dataset")
plt.show()
```

```python
labels = (diagnoses.keys())
values = (diagnoses.values())
plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=["red", "green"])
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.title("Distribution of Diagnosis in a dataset")
plt.show()
```

```python
## Create scatter plots to explore the relationships between 'Radius (mean)' and 'Texture (mean)', 'Area (mean)' and 
##'Smoothness (mean)', etc. Use different colors or shapes to differentiate between malignant and benign cases.
```

```python
sns.scatterplot(x=df["Radius (mean)"],y=df["Texture (mean)"],color='red', label='Malignant') 
sns.scatterplot(x=df["Radius (mean)"],y=df["Texture (mean)"],color='blue', label='Benign') 
## this analysis indicates a negligible presence of cancer cases within the relationship between 
## radius and texture, as benign cases outnumbered malignant cases.
```

```python
sns.scatterplot(x=df["Area (mean)"],y=df["Smoothness (mean)"],color='green', label='Malignant')
sns.scatterplot(x=df["Area (mean)"],y=df["Smoothness (mean)"],color='purple', label='Benign')
## this analysis indicates a negligible presence of cancer cases within the relationship between 
## Area and Smoothness, as benign cases outnumbered malignant cases.
```

```python
## Generate a heatmap to visualize the correlation matrix between different features. Identify which features have a strong
## correlation with each other and with the 'Diagnosis' column.
```

```python
df2=df[["Radius (mean)","Texture (mean)","Perimeter (mean)","Smoothness (mean)"]]
corr=df2.corr()
corr
```

```python
sns.heatmap(corr,annot=True,fmt="0.2g")
```

```python
df2=df[["Compactness (worst)","Symmetry (worst)","Smoothness (worst)","Fractal dimension (worst)"]]
corr=df2.corr()
corr
```

```python
sns.heatmap(corr,annot=True,fmt="0.2g")
```
