```python
fig= plt.figure(figsize=(20,20))
axes=fig.add_axes([0.1,0.1,0.7,0.4])
plt.style.use("ggplot")
axes.plot(df2["year"],df2["Agriculture"],ls="--",lw=3.5,marker="o",markevery=3,mec="green",mfc="purple",color= "green", label="Agric")
axes.plot(df2["year"],df2["Education"],ls="--",lw=3.5,marker="o",markevery=3,mec="blue",mfc="purple",color="blue", label = "Educ")
axes.plot(df2["year"],df2["CrudepetroleumAndNaturalGas"],ls="--",lw=3.5,marker="o",markevery=3,mec="black",mfc="purple",color = "black", label = "CRUDE")
axes.plot(df2["year"],df2["InformationAndCommunication"],ls="--",lw=3.5,marker="o",markevery=3,mec="yellow",mfc="purple", color = "yellow", label="ICT")
axes.set_xlabel("YEAR",color = "#833da8", fontsize = 25)
axes.set_ylabel("GDP", color = "#13806d", fontsize = 25)
plt.title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color = "navy",fontsize = 25)
plt.legend()
plt.show()
```

```python
fig, axes=plt.subplots(2,2, figsize =(15,10))
axes[0,0].bar(df2["year"],df2["Agriculture"], color ="green")
axes[0,0].set_xlabel("YEAR",color="navy", fontsize=12)
axes[0,0].set_ylabel("Agriculture",color="navy", fontsize=12) 
axes[0,0].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color = "navy",fontsize = 12)  
              
axes[1,0].step(df2["year"],df2["Education"], color="yellow")
axes[1,0].set_xlabel("YEAR",color="navy", fontsize=12)
axes[1,0].set_ylabel("Education",color="navy", fontsize=12) 
axes[1,0].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color = "navy",fontsize = 12)  
                            
axes[0,1].scatter(df2["year"],df2["CrudepetroleumAndNaturalGas"], color="black")
axes[0,1].set_xlabel("YEAR",color="navy", fontsize=12)
axes[0,1].set_ylabel("CrudepetroleumAndNaturalGas",color="navy", fontsize=12) 
axes[0,1].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color = "navy",fontsize = 12)  
                            
axes[1,1].hist(df2["InformationAndCommunication"], bins="auto", color="blue")
axes[1,1].set_xlabel("YEAR",color="navy", fontsize=12)
axes[1,1].set_ylabel("InformationAndCommunication",color = "navy", fontsize=12) 
axes[1,1].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color = "navy",fontsize = 12)  
     
plt.show()
```
